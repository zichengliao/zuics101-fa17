```python
import numpy as np
n = 10
start = 0.0
end = 1.0
a = -9.8
t = np.linspace(start, end, n+1)

y0 = 1.0 + a/2*t**2
for i in range(1,n+1):
    if y0[i] <= 0:
        y0[i] = 0
```

```python
import matplotlib.pyplot as plt
plt.plot(t,y0,'b--')
plt.xlabel('time')
plt.ylabel('height')
plt.show()
```

```python
import numpy as np

# parameters of simulation
n = 1000
start = 0.0
end = 1.0
a = -9.8

# state variable initialization
t = np.linspace(start, end, n+1)  #time, s
v = np.zeros(n+1)
y = np.zeros(n+1)
y[0] = 1.0

for i in range(1,n+1):
    v[i] = v[i-1] + a*(t[i]-t[i-1])
    y[i] = y[i-1] + v[i]*(t[i]-t[i-1])
    if y[i] <= 0:
        v[i] = -0.5*v[i]  #0
        y[i] = -y[i]  #0
```

```python
import matplotlib.pyplot as plt
plt.plot(t,y,'r--')
plt.xlabel('time')
plt.ylabel('height')
plt.title('Falling trajectory')
#plt.show()

plt.plot(t,y0,'b--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('populationHeight.csv')
plt.hist(data, 50)
plt.xlim(135, 200)
plt.show()
```

```python
mean = data.mean()
std = data.std()
print([mean, std])
```

```python
from math import pi
x = np.linspace(140,200,500)
y = 1/((2*pi)**0.5*std)*np.exp(-(x-mean)**2/(2*std**2))
```

```python
plt.hist(data, 50)
plt.plot(x,y*10000,'r-')
plt.xlabel('height')
plt.title('fit Gaussian to histogram')
plt.xlim(135, 200)
plt.show()
```

```python
import matplotlib.pyplot as plt

xs = list(range(4))
ys = [4.5, 6, 1.2, 5.4]
```

```python
print(xs)
print(ys)
```

```python
plt.plot(xs, ys, 'b--', linewidth=3.0)
plt.xlabel('domain')
plt.ylabel('range')
plt.title('plot of my data')
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)
print(x)
print(y)
```

```python
plt.plot(x, y, 'k--')
plt.xlim(0, 2*np.pi)
plt.show()
```

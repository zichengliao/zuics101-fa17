```python
for i in 'ABCD':
   for j in 'ABCD':
      if i == j:
        continue
      print( i, j )
```

```python
from itertools import permutations
for triplet in permutations( 'ABCD',3 ):
   print( triplet )
```

```python
for triplet in permutations( 'ABCD',4 ):
   print( triplet )
```

```python
for triplet in permutations( 'ABCD',5 ):
   print( triplet )
```

```python
for triplet in permutations(list(range(10)), 3):
    print( triplet )
```

```python
for i, item in enumerate(permutations(list(range(10)), 3)):
    print(i, item)
```
